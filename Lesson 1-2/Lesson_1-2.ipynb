### Задача с треугольниками
Далее приведена ф-ция, генерирующая входные данные для данной задачи. N - число точек

```python
def generate_points(N):
    x = 2*np.random.rand(N)-1
    y = (np.random.randint(2, size = N)*2-1) * np.sqrt(1-x**2) 
    return x, y
```

```python
data_x, data_y = generate_points(32)
plt.scatter(data_x, data_y, marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
def alg1(data_x, data_y):
    N = data_x.size;
    S_max = 0;
    i_max = []
    
    for i1 in range(N-2):
        for i2 in range(i1+1, N-1):
            for i3 in range(i2+1, N):
                a = np.sqrt((data_x[[i1, i2, i3]] - data_x[[i2, i3, i1]])**2 + \
                    (data_y[[i1, i2, i3]] - data_y[[i2, i3, i1]])**2)
                S = a[0] * a[1] * a[2] / 4;
                if(S > S_max):
                    S_max = S
                    i_max = [i1, i2, i3]
                    
    return i_max
```

```python
i_max = alg1(data_x, data_y)
```

```python
x = np.append(data_x[i_max], data_x[i_max][0])
y = np.append(data_y[i_max], data_y[i_max][0])
plt.fill(x, y)
plt.scatter(data_x, data_y, marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
